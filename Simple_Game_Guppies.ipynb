## Introduction to the Guppies Game

Guppies is a simple betting game where:
* a player starts with 100 dollars
* a player may bet as much of their money as they like every turn
* the player places a bet and the computer checks that it is valid
* the computer selects and announces one random number between 1 and 10
* the player guesses if the next random number will be higher or lower than the first number
* the second random number is then selected and announced
* if the player's guess was correct, then the player earns money equivalent to their bet; otherwise, they lose the money they bet
* the game continues until the player loses all of their money or earns 1000 dollars.
